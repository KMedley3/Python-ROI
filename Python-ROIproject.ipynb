```python

class roi():
    
    # create a method to pass income, expenses, and total investment to calculate roi
    def __init__(self, property_name):
        self.property_name = property_name
            
        
    # create a function to ask for input pertaining to income, add all inputs and save to a variable to pass to other functions
    def income(self):
        rental_income = int(input("What is your rental income per month? Numbers only: "))
        storage = int(input("How much per month do you make from storage? Numbers only: "))
        Laundry = int(input("How much per month do you make from laundry? Numbers only: "))
        Miscellaneous = int(input("What's the total amount for all other monthly expenses? Numbers only: "))
        
        self.total_income = rental_income + storage + Laundry + Miscellaneous
        
#  # create a function for to ask for input pertaining to all expenses and add them all up. Save the variable to pass
# #  to other functions

    def expenses(self):
        taxes = int(input("How much does your property cost you in taxes? Numbers only: "))
        insurance = int(input("How much do you pay in insurance for this property? Numbers only: "))
        utilities = int(input("How much do you pay in utilities for this property? Numbers only: "))
        hoa = int(input("If you have an HOA fee, how much is it? If you don't, enter 0. Numbers only: "))
        lawn_care = int(input("How much do you spend in lawn care for your property? Numbers only: "))
        vacancy = int(input("How much are you saving per month for vacancy? Numbers only: "))
        repairs = int(input("How much do you save per month for repairs? Numbers only: "))
        capEx = int(input("How much do you save per month for capital expenditure? Numbers only: "))
        manager = int(input("How much do you pay your property manager? If you don't have one enter 0. Numbers only: "))
        mortgage = int(input("How much is your mortgage payment? Numbers only: "))
        
        self.total_expenses = taxes + insurance + hoa + utilities + lawn_care + vacancy + repairs + capEx + manager + mortgage
        print(self.total_expenses)
        print(self.total_income)
        
    def get_cashflow(self):
        monthly_cashflow = int(self.total_income) - int(self.total_expenses)
        print(f"Your total cashflow is {monthly_cashflow}")
        self.annual_income = int(monthly_cashflow) * 12
        print(f"Your annual income is {self.annual_income}")
        
        
    def investment(self):
        down_payment = int(input("What was the down payment for this property? "))
        closing_cost = int(input("What was the closing cost for this property? "))
        rehab = int(input("Was there any rehab for this property? If none, enter 0. "))
        other = int(input("Are there any other expenses? If none, enter 0. "))
        
        self.total_investment = down_payment + closing_cost + rehab + other
        print(self.total_investment)
        
    
    def calc_roi(self):
        return_invest = (self.annual_income / self.total_investment) * 100
        print(return_invest)
    
        
        
    

            

myRoi = roi(input("What is the name of this property? "))           
while True:
    answer = input("What would you like to calculate? Income/Expenses/Cash flow/total investment/Roi")
    if answer == 'Income':
        myRoi.income()
    elif answer == 'Expenses':
        myRoi.expenses()
    elif answer == 'Cash flow':
        myRoi.get_cashflow()
    elif answer == 'total investment':
        myRoi.investment()
    elif answer == 'Roi':
        myRoi.calc_roi()
    else:
        break
    


    
    
    
    
```
